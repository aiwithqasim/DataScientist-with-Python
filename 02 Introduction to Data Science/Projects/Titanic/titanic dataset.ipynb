### Summary

```python
d = { 0: 'Not survived',
    1: 'Survived'}
```

```python
data['stats'] = data.Survived.map(d)
data
```

```python
def decrypt(x):
    if x == 0: return 'Not Survived'
    elif x == 1 : return 'Survived' 
    else: return 'Invalid'
```

```python
data['with_apply'] = data.Survived.apply(decrypt)
```

```python
data.head()
```

```python
def endcode(x):
    if x == 'Not Survived': return 0
    elif x == 'Survived' : return  1
    else: return 'Invalid'
```

```python
data['endcode_survived'] = data.with_apply.apply(endcode) 
```

```python
data.head()
```

```python
def endcode_gender(x):
    if x == 'male': return 0
    elif x == 'female' : return  1
    else: return 'Invalid'
```

```python
data['gender_encode'] = data.Sex.apply(endcode_gender) 
```
